This is a classic problem of time series (TS) forecast. There are many methods. Exponential Smoothing, ARIMA, reocurring neural network. Due to the limitation of time, I only use exponetial smoothing, and built it in R.

## Load libraries

```r
#install.packages("forecast")
library("forecast")
library("graphics")
```

## Data preprocessing

```r
dataset = read.csv('/Users/meliu/Downloads/sample_data.csv')
dataset = dataset[c(27)]
ts <- ts(dataset, start=c(2006, 10), end=c(2015, 12), frequency=12) 
ts
```

## Built model and forecast

```r
fit1 <- hw(ts,seasonal="additive")
fit2 <- hw(ts,seasonal="multiplicative")

plot(fit2,ylab="Car Residual Value",
     plot.conf=FALSE, type="o", fcol="white", xlab="Year")
lines(fitted(fit1), col="red", lty=2)
lines(fitted(fit2), col="green", lty=2)
lines(fit1$mean, type="o", col="red")
lines(fit2$mean, type="o", col="green")
legend("topright",lty=1, pch=1, col=1:3, 
  c("data","Holt Winters' Additive","Holt Winters' Multiplicative"))
```

## Forecasting results

```r
states <- cbind(fit1$model$states[,1:3],fit2$model$states[,1:3])
colnames(states) <- c("level","slope","seasonal","level","slope","seasonal")
plot(states, xlab="Year")
fitted(fit1)
fit1$mean
fitted(fit2)
fit2$mean
```
